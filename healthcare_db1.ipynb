```python
import mysql.connector
import pandas as pd

def fetch_data(query, cursor):
    cursor.execute(query)
    return pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])

try:
    # Establish database connection
    connection = mysql.connector.connect(
        user='root',
        password='iloveyou0807',
        host='localhost',
        database='healthcare_db'
    )
    print("Connection established successfully.")
    
    # Create a cursor object to execute SQL queries
    cursor = connection.cursor(buffered=True)

    # Fetch data from all tables and store in DataFrames
    Patients_data = fetch_data('SELECT * FROM Patients', cursor)
    Specialties_data = fetch_data('SELECT * FROM Specialties', cursor)
    Doctors_data = fetch_data('SELECT * FROM Doctors', cursor)
    Appointments_data = fetch_data('SELECT * FROM Appointments', cursor)
    Prescriptions_data = fetch_data('SELECT * FROM Prescriptions', cursor)

    # Check for missing values in the DataFrame and sum them up column-wise
    print("Patients_data\n----------")
    print(Patients_data.isnull().sum())
    
    print("Specialties_data\n----------")
    print(Specialties_data.isnull().sum())
    
    print("Doctors_data\n----------")
    print(Doctors_data.isnull().sum())
    
    print("Appointments_data\n----------")
    print(Appointments_data.isnull().sum())
    
    print("Prescriptions_data\n----------")
    print(Prescriptions_data.isnull().sum())

except mysql.connector.Error as err:
    print(f"Error: {err}")
finally:
    if 'cursor' in locals():
        cursor.close()
    if 'connection' in locals() and connection.is_connected():
        connection.close()
```

## Step 5 : Deleting NULL values or incomplete tables

```python
import mysql.connector

# Connect to the database
connection = mysql.connector.connect(
    user='root',
    password='iloveyou0807',
    host='localhost',
    database='healthcare_db'
)
cursor = connection.cursor(buffered=True)

# Fill missing values in the 'Specialties' table
cursor.execute("UPDATE Specialties SET specialty_name = 'Unknown' WHERE specialty_name IS NULL")

# Drop rows with missing values in the 'Appointments' table
cursor.execute("DELETE FROM Appointments WHERE patient_id IS NULL OR doctor_id IS NULL OR appointment_date IS NULL")

# Commit the changes
connection.commit()

# Close the cursor and connection
cursor.close()
connection.close()

print("All null values deleted.\n")
```

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt

try:
    # Establish database connection
    connection = mysql.connector.connect(
        user='root',
        password='iloveyou0807',
        host='localhost',
        database='healthcare_db'
    )
    print("Connection established successfully.")
    
    # Create a cursor object to execute SQL queries
    cursor = connection.cursor(buffered=True)

    # Query data from the 'Doctors' table
    cursor.execute('SELECT * FROM Doctors')
    Doctors_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])
    
    # Print the columns of Doctors_data to verify
    print("Doctors_data columns:", Doctors_data.columns)

    # Query data from the 'Appointments' table
    cursor.execute('SELECT * FROM Appointments')
    Appointments_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])
    
    # Print the columns of Appointments_data to verify
    print("Appointments_data columns:", Appointments_data.columns)

    # Grouping by doctor_id and counting appointments
    doctor_appointments = Appointments_data.groupby('doctor_id').size()

    # Check for existence of 'doctor_id' and 'doctor_name' columns
    if 'doctor_id' in Doctors_data.columns and 'doctor_name' in Doctors_data.columns:
        # Mapping doctor IDs to doctor names
        doctor_appointments.index = doctor_appointments.index.map(Doctors_data.set_index('doctor_id')['doctor_name'])
    else:
        print("The required columns ('doctor_id' and 'doctor_name') are not present in the Doctors_data DataFrame.")
    
    # Plotting
    plt.figure(figsize=(10, 6))
    doctor_appointments.plot(kind='bar', color='lightcoral')
    plt.title('Number of Appointments per Doctor')
    plt.xlabel('Doctor')
    plt.ylabel('Number of Appointments')
    plt.xticks(rotation=45, ha='right')  # Adjust rotation and alignment of x-labels
    plt.grid(axis='y')
    plt.show()

except mysql.connector.Error as err:
    print(f"Error: {err}")
finally:
    if 'cursor' in locals():
        cursor.close()
    if 'connection' in locals() and connection.is_connected():
        connection.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mysql.connector

try:
    # Establish database connection
    connection = mysql.connector.connect(
        user='root',
        password='iloveyou0807',
        host='localhost',
        database='healthcare_db'
    )
    print("Connection established successfully.")
    
    # Create a cursor object to execute SQL queries
    cursor = connection.cursor(buffered=True)

    # Query data from the 'Appointments' table
    cursor.execute('SELECT * FROM Appointments')
    
    # After fetching data from the database, we are storing it into Pandas DataFrame
    Appointments_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])
    
    # Print the first few rows of the DataFrame to verify the data
    print(Appointments_data.head())

    # Convert appointment_date to datetime if it's not already
    Appointments_data['appointment_date'] = pd.to_datetime(Appointments_data['appointment_date'])

    # Extract month and year from appointment_date
    Appointments_data['year_month'] = Appointments_data['appointment_date'].dt.to_period('M').astype(str)

    # Group by year_month and count the number of appointments
    monthly_appointments = Appointments_data.groupby('year_month').size().reset_index(name='num_appointments')

    # Plot the data
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=monthly_appointments, x='year_month', y='num_appointments')
    plt.title('Number of Appointments by Month')
    plt.xlabel('Month')
    plt.ylabel('Number of Appointments')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

except mysql.connector.Error as err:
    print(f"Error: {err}")
finally:
    if 'cursor' in locals():
        cursor.close()
    if 'connection' in locals() and connection.is_connected():
        connection.close()
```

```python
from datetime import datetime

try:
    # Establish database connection
    connection = mysql.connector.connect(
        user='root',
        password='iloveyou0807',
        host='localhost',
        database='healthcare_db'
    )
    print("Connection established successfully.")
    
    # Create a cursor object to execute SQL queries
    cursor = connection.cursor(buffered=True)

    # Query to get the appointment dates
    query = '''
    SELECT appointment_date
    FROM appointments
    '''
    
    cursor.execute(query)
    
    # Storing the results into a Pandas DataFrame
    appointment_dates_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])
    
    # Calculate the wait times
    current_date = datetime.now()
    appointment_dates_data['appointment_date'] = pd.to_datetime(appointment_dates_data['appointment_date'])
    appointment_dates_data['wait_time_days'] = (appointment_dates_data['appointment_date'] - current_date).dt.days

    # Plotting the wait times
    plt.figure(figsize=(10, 6))
    sns.histplot(appointment_dates_data['wait_time_days'], bins=30, kde=True)
    plt.title('Distribution of Appointment Wait Times')
    plt.xlabel('Wait Time (days)')
    plt.ylabel('Frequency')
    plt.show()

except mysql.connector.Error as err:
    print(f"Error: {err}")
finally:
    if 'cursor' in locals():
        cursor.close()
    if 'connection' in locals() and connection.is_connected():
        connection.close()
```
